```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras import layers, models
import joblib
```

```python
data=pd.read_csv('/content/heart (2).csv')
print("dataset succesfully loaded",data.shape)
```

```python
data= data.dropna()
# Handle missing entries and drop duplicate records
data = data.drop_duplicates()
```

```python
# Clean structural text anomalies in 'thal' column
data['thal'] = data['thal'].astype(str).str.strip().str.lower()
print("--- Data Preparation ---")
print(f"Dataset successfully processed. Operational Shape: {data.shape}\n")

# *******************************************
# LABEL ENCODING
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
label_encoder = LabelEncoder()

data['thal'] = label_encoder.fit_transform(data['thal'])

x=data.drop('target',axis=1)
y=data['target']


```

```python
#now saving the model
joblib.dump(label_encoder, 'label_encoder.pkl')
feature_columns = x.columns.tolist()
joblib.dump(feature_columns, 'feature_columns.pkl')
# Save the exact order of features for the prediction program
feature_columns = x.columns.tolist()
joblib.dump(feature_columns, 'feature_columns.pkl')

# 4. TRAIN-TEST SPLIT & MODEL TRAINING
# ***********-----@@@@!!!!!!

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Save the fitted scaler framework and label encoder to Colab storage
joblib.dump(scaler, "ANN_scaler.pkl")
joblib.dump(label_encoder, "label_encoder_heart.pkl")

print("Data preprocessing completed. Scaler and Encoder saved successfully.")
```

```python
# =====================================================================
# CELL 2: ANN MODEL TRAINING & EVALUATION PIPELINE
# =====================================================================
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Build Artificial Neural Network (ANN) with Multiple Hidden Layers
model = Sequential([
    Dense(16, activation='relu', input_shape=(x_train_scaled.shape[1],)), # Hidden Layer 1
    Dense(8, activation='relu'),                                      # Hidden Layer 2
    Dense(4, activation='relu'),                                      # Hidden Layer 3
    Dense(1, activation='sigmoid')                                    # Output Layer (Sigmoid)
])

# Compiles with binary_crossentropy, which matches a single-neuron Sigmoid configuration
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print("--- Model Structural Summary ---")
model.summary()

print("\n--- Starting Neural Network Training Loops ---")

model.fit(x_train_scaled, y_train, epochs=50)

# ***********@@@@@@@@@@@!!!!!!!!!!!
# 5. EVALUATION
# ^^^^^^***********************
loss, accuracy = model.evaluate(x_test_scaled, y_test,)

print("\n--- Model Evaluation Metrics ---")
print(f"Test Set Binary Crossentropy Loss: {loss:<.5f}")
print(f"Test Set Overall Classification Accuracy: {accuracy * 100:.2f}%")

#**************
# 6. SAVE ARTIFACTS
#$$$$$$$$$$$$$$%$$$$$$$$$$$
model.save("heart_disease_ANN_model.keras")
print("Model trained and saved successfully.")
```

```python
##### TASK 6 TAKING DETAILS FROM PATIENT

import numpy as np
import joblib
from tensorflow.keras.models import load_model

# Load your exact saved assets from Cell 1
scaler = joblib.load("ANN_scaler.pkl")
model = load_model("heart_disease_ANN_model.keras")
le = joblib.load("label_encoder_heart.pkl")

print("Enter Patient Details #####*****&&&&&+++++++")

age = float(input("Age: "))
sex = int(input("Sex (1=Male, 0=Female): "))
cp = int(input("Chest Pain Type (1, 2, 3, or 4): "))
trestbps = float(input("Resting Blood Pressure: "))
chol = float(input("Cholesterol: "))
fbs = int(input("Fasting Blood Sugar > 120 (1=Yes, 0=No): "))
restecg = int(input("Rest ECG (0, 1, or 2): "))
thalach = float(input("Maximum Heart Rate Achieved: "))
exang = int(input("Exercise Induced Angina (1=Yes, 0=No): "))
oldpeak = float(input("Oldpeak ST Depression: "))
slope = int(input("Slope (1, 2, or 3): "))
ca = int(input("Number of Major Vessels (0-3): "))

# Request text input to match your dataset's LabelEncoder mapping rules
thal_input = input("Thal (normal / fixed / reversible): ").strip().lower()

# Apply your saved LabelEncoder to convert the text input into the correct code
if thal_input in le.classes_:
    thal_encoded = le.transform([thal_input])[0]
else:
    print(f"вҡ пёҸ Value '{thal_input}' not recognized. Defaulting to 'normal'.")
    thal_encoded = le.transform(['normal'])[0]

# Build the patient array using your encoded text value
patient = np.array([[age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal_encoded]])
patient_df = pd.DataFrame(patient, columns=feature_columns)

patient_scaled = scaler.transform(patient_df)

probability = model.predict(patient_scaled, verbose=0)[0][0]

print("\nPrediction Probability:", round(probability, 4))

if probability >= 0.5:
    print("Heart Disease Detected")
else:
    print("No Heart Disease")
```
